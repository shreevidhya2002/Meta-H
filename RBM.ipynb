```python
import numpy as np
import pandas as pd

class RBM:
    def __init__(self, num_visible, num_hidden):
        self.num_visible = num_visible
        self.num_hidden = num_hidden
        self.weights = np.random.randn(num_visible, num_hidden) / np.sqrt(num_visible)
        self.visible_bias = np.zeros(num_visible)
        self.hidden_bias = np.zeros(num_hidden)

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def contrastive_divergence(self, v0, vk, ph0, phk):
        self.weights += np.outer(v0, ph0) - np.outer(vk, phk)
        self.visible_bias += np.asarray(v0, dtype=np.float32) - np.asarray(vk, dtype=np.float32)
        self.hidden_bias += np.asarray(ph0, dtype=np.float32) - np.asarray(phk, dtype=np.float32)

    def sample_hidden(self, v):
        activation = np.dot(v, self.weights) + self.hidden_bias
        p_h_given_v = self.sigmoid(activation)
        return p_h_given_v > np.random.rand(self.num_hidden)

    def sample_visible(self, h):
        activation = np.dot(h, self.weights.T) + self.visible_bias
        p_v_given_h = self.sigmoid(activation)
        return p_v_given_h > np.random.rand(self.num_visible)

    def train(self, data, max_epochs=1000, learning_rate=0.1):
        num_examples = data.shape[0]
        for epoch in range(max_epochs):
            for i in range(num_examples):
                v0 = data[i]
                ph0 = self.sample_hidden(v0)
                vk = v0  # Initialize vk before the loop
                for k in range(10):  # Gibbs sampling
                    hk = self.sample_visible(self.sample_hidden(vk))
                    vk = self.sample_visible(self.sample_hidden(hk))
                phk = self.sample_hidden(vk)
                self.contrastive_divergence(v0, vk, ph0, phk)


```

```python
def RBM_Foraging_Optimization(X, y):
    X_array = X.values  # Convert DataFrame to NumPy array
    num_visible = X_array.shape[1]
    num_hidden = 100  # Adjust according to your problem
    rbm = RBM(num_visible, num_hidden)
    rbm.train(X_array)  # Pass NumPy array instead of DataFrame
    return rbm
```
