```python
def mean_squared_error(actual_values, predicted_values):
    """
    Calculate Mean Squared Error (MSE) between actual and predicted values.
    
    Parameters:
        actual_values (list): List of actual values.
        predicted_values (list): List of predicted values.
    
    Returns:
        float: Mean Squared Error (MSE) between actual and predicted values.
    """
    if len(actual_values) != len(predicted_values):
        raise ValueError("Input lists must have the same length.")
    
    squared_errors = [(actual - predicted) ** 2 for actual, predicted in zip(actual_values, predicted_values)]
    mse = sum(squared_errors) / len(actual_values)
    return mse

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.metrics.pairwise import cosine_similarity
import pickle

def predict_attack(csv_file_path, model_file_path='cyber_model.h5', scaler_file_path='standard_scaler.pkl', train_data_file_path='scaled_train_data.pkl'):
    # Load the model
    loaded_model = load_model(model_file_path)
    
    # Load the standard scaler
    with open(scaler_file_path, 'rb') as f:
        scaler = pickle.load(f)
    
    # Load the test data from the CSV file
    test_data = pd.read_csv(csv_file_path, header=None)  # Assuming no header in the CSV file
    
    # Scale the test data using the same scaler used during training
    scaled_test_data = scaler.transform(test_data)
    
    # Reshape input data to have an additional dimension for channels
    scaled_test_data = scaled_test_data.reshape(scaled_test_data.shape[0], scaled_test_data.shape[1], 1)
    
    # Use the loaded model to make predictions
    predictions = loaded_model.predict(scaled_test_data)
    
    # Calculate similarity scores between input data and trained data
    with open(train_data_file_path, 'rb') as f:
        X_train_scaled = pickle.load(f)
    
    similarities = cosine_similarity(scaled_test_data.reshape(scaled_test_data.shape[0], -1), X_train_scaled.reshape(X_train_scaled.shape[0], -1))
    most_similar_indices = np.argmax(similarities, axis=1)
    max_similarities = np.max(similarities, axis=1)
    
    # Calculate the percentage of matching features
    matching_percentage = max_similarities * 100
    
    # Convert predictions to class labels
    class_labels = np.argmax(predictions, axis=1)
    
    # Get the predicted class
    predicted_class = class_labels[0]  # Assuming only one prediction is made
    
    # Print the result
    print("Predicted labels:", class_labels)
    print("Most similar training indices:", most_similar_indices)
    print("Max similarity scores:", max_similarities)
    print("The input is matching the trained row by {:.2f}%, so the predicted class is {}".format(matching_percentage[0], predicted_class))

# Example usage:
csv_file_path = 'C:/Users/shree/Downloads/cyber security/Scenorios/1.csv'
predict_attack(csv_file_path)
```
